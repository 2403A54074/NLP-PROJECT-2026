```python
!pip install googletrans==3.1.0a0

import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from googleapiclient.discovery import build
from googletrans import Translator
from IPython.display import display, clear_output
import re

# ================= CONFIG =================
API_KEY = "AIzaSyCjp89ALA8-Dy2diShelcYnty-JV1Nay44"   # ❂ Put your YouTube API key here
translator = Translator()

# ================= UI =================
url_input = widgets.Text(description='YouTube URL:')

limit_input = widgets.IntSlider(
    value=200,
    min=50,
    max=5000,
    step=50,
    description='Limit:'
)

order_dropdown = widgets.Dropdown(
    options=[('Top Comments','relevance'), ('Latest','time')],
    value='relevance',
    description='Order:'
)

lang_dropdown = widgets.Dropdown(
    options=[
        ('English','en'), ('Hindi','hi'), ('Telugu','te'),
        ('Tamil','ta'), ('French','fr'), ('Spanish','es'),
        ('Arabic','ar'), ('German','de'),
        ('Chinese','zh-cn'), ('Japanese','ja'),
        ('Korean','ko'), ('Russian','ru')
    ],
    value='hi',
    description='Convert To:'
)

load_btn = widgets.Button(description="Load Comments")
convert_btn = widgets.Button(description="Translate All")
output = widgets.Output()

display(url_input, limit_input, order_dropdown, lang_dropdown, load_btn, convert_btn, output)

# ================= FUNCTIONS =================

# Extract video ID
def extract_video_id(url):
    # Updated regex to correctly extract the 11-character video ID from common YouTube URL formats.
    regex = r'(?:v=|youtu\.be\/|embed\/|watch\?v=)([0-9A-Za-z_-]{11})'
    match = re.search(regex, url)
    return match.group(1) if match else None

# Fetch comments
def get_comments(video_id, limit, order):
    youtube = build('youtube', 'v3', developerKey=API_KEY)
    comments = []
    next_page_token = None

    while len(comments) < limit:
        request = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            maxResults=100,
            order=order,
            pageToken=next_page_token
        )
        response = request.execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comments.append(comment)

            if len(comments) >= limit:
                break

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return comments

comments_data = []

# ================= LOAD COMMENTS =================
def load_comments(b):
    global comments_data
    with output:
        clear_output()

        video_id = extract_video_id(url_input.value)
        if not video_id:
            print("❌ Invalid YouTube URL")
            return

        print(f"⏳ Loading {limit_input.value} comments...")
        comments_data = get_comments(video_id, limit_input.value, order_dropdown.value)

        print(f"✅ Loaded {len(comments_data)} comments successfully!")

load_btn.on_click(load_comments)

# ================= TRANSLATE COMMENTS =================
def translate_comments(b):
    with output:
        clear_output()

        if not comments_data:
            print("❌ Please load comments first")
            return

        target_lang = lang_dropdown.value
        data = []

        print("⏳ Translating comments... (Auto-detect mode)")

        for i, c in enumerate(comments_data):
            try:
                # Google-style auto detect + translate
                translated = translator.translate(c, dest=target_lang).text
            except:
                translated = "Translation Error"

            data.append([c, translated])

        df = pd.DataFrame(data, columns=["Original Comment", "Translated Comment"])

        display(df)

        print("\n✅ Translation Completed!")

translate_btn = convert_btn
translate_btn.on_click(translate_comments)
```
